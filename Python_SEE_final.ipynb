using both recursion and iteration. Compare their time complexity.

```python
for i in range(1,51):
  if(i%5==0):
    continue
  print(i)
```

```python
import time

def fibo_iter(n):
  fib_series = [0,1]
  for i in range(n-2):
    fib_series.append(fib_series[-1]+fib_series[-2])
  return fib_series[:n]

def fibo_recur(n):
  if n<=1:
    return n
  return fibo_recur(n-1)+fibo_recur(n-2)

n = int(input("Enter a number : "))

start = time.time()
print("Iterative fibonacci : ",fibo_iter(n))
print("Time taken : ",time.time()-start)
start = time.time()
print("Recursive fibonacci : ",[fibo_recur(i) for i in range(n)])
print("Time taken : ",time.time()-start)
```

i. Write a Python decorator that prints "Before calling function" and
"After calling function" around the execution of any function.

ii. Write a decorator that converts the output of a function to uppercase.

```python
def before_after(func):
  def wrapper():
    print("Before calling function")
    func()
    print("After calling function")
  return wrapper

@before_after
def say_hello():
  print("Hello !")

say_hello()
```

```python
def uppercase_function(func):
  def wrapper():
    return func().upper()
  return wrapper

@uppercase_function
def say_hello():
  return "hello !"

print(say_hello())
```

BankAccount with private attributes __balance and methods to

ii. Write a Python program that demonstrates abstraction using an
interface (abstract class) for different payment methods like
CreditCard, PayPal, and Bitcoin.

```python
class BankAccount:
  def __init__(self,balance):
    self.__balance = balance
  def deposit(self,amount):
    self.__balance += amount
  def withdraw(self,amount):
    self.__balance -= amount
  def get_balance(self):
    print("Current balance is ",self.__balance)

myaccount = BankAccount(1000)
myaccount.get_balance()
myaccount.deposit(500)
myaccount.get_balance()
myaccount.withdraw(200)
myaccount.get_balance()
```

```python
from abc import ABC, abstractmethod

class PaymentMethod(ABC):
  @abstractmethod
  def pay(self,amount):
    pass
class CreditCard(PaymentMethod):
  def pay(self,amount):
    print("Paid Rs.",amount,"using Credit Card")
class PayPal(PaymentMethod):
  def pay(self,amount):
    print("Paid Rs.",amount,"using PayPal")
class Bitcoin(PaymentMethod):
  def pay(self,amount):
    print("Paid Rs.",amount,"using Bitcoin")

#usage
for method in [CreditCard(),PayPal(),Bitcoin()]:
  method.pay(100)
```

i. Create a class Laptop with private attributes for price and model, and
public methods to access and modify them safely.

ii. Create an abstract class Shape with an abstract method
calculate_area(), then implement Circle and Square classes.

```python
class laptop:
  def __init__(self,model,price):
    self.__model = model  #private
    self.__price = price

  def getmodel(self):    #public
    return self.__model

  def getprice(self):
    return self.__price

  def setmodel(self,model):
    self.__model = model

  def setprice(self,price):
    if price>0:
      self.__price = price

l = laptop("Dell",50000)
print("Model : ",l.getmodel())
print("Price : ",l.getprice())
l.setprice(69000)
l.setmodel("HP Victus")
print("Model : ",l.getmodel())
print("Price : ",l.getprice())
```

```python
from abc import ABC, abstractmethod

class shape(ABC):
  @abstractmethod
  def calculate_area(self):
    pass

class circle(shape):
  def __init__(self,radius):
    self.radius = radius
  def calculate_area(self):
    return 3.14*self.radius*self.radius

class square(shape):
  def __init__(self,side):
    self.side = side
  def calculate_area(self):
    return self.side*self.side

C = circle(7)
S = square(5)
print("Area of circle is : ",C.calculate_area())
print("Area of square is : ",S.calculate_area())
```
